```python
import findspark
findspark.init("C:\spark")

from pyspark.sql import SparkSession
from pyspark.conf import SparkConf

spark = SparkSession.builder. \
master("local[4]"). \
appName("Df-giriş"). \
config("spark.driver.memory","2g"). \
config("spark.executor.memory","4g"). \
getOrCreate()
```

```python
adult_train_df = spark.read \
.option("header","True") \
.option("inferSchema","True") \
.option("sep",",") \
.csv(r"C:\Users\MERT\Desktop\spark-adultdata\Data\adult.data") \
```

```python
adult_test_df = spark.read \
.option("header","True") \
.option("inferSchema","True") \
.option("sep",",") \
.csv(r"C:\Users\MERT\Desktop\spark-adultdata\Data\adult.test") \
```

# Let's look Categorical Vars

### Genelde categoric verileri incelerken group by yaparsan iyi bilgiler veriyo

## 1-workclass

```python
## Sql sorgularını kullancagım ıcın pyspark sql i import ediyorm
from pyspark.sql import functions as F
```

```python
adult_df.groupBy(F.col("workclass")) \
.agg({"*":"count"}) \
.toPandas().head(10)
```
